```python
%matplotlib inline
from __future__ import print_function
import tifffile as tiff
import matplotlib.pyplot as plt

img_trn = tiff.imread('./train-volume.tif')
msk_trn = tiff.imread('./train-labels.tif')

for img,msk in zip(img_trn, msk_trn):
    fig,_ = plt.subplots(nrows=1, ncols=2)
    fig.axes[0].imshow(img, cmap='gray')
    fig.axes[1].imshow(msk, cmap='gray')
    plt.show()
```
